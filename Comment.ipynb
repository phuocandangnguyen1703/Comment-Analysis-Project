```python
sentences = nltk.sent_tokenize(contents_parsed[3])
print(sentences)
```

```python
from gensim.models import KeyedVectors 

w2v = KeyedVectors.load_word2vec_format("/content/drive/MyDrive/Colab Notebooks/Khoa luan/2.Comment Project/vi.vec")
```

```python
vocab = w2v.key_to_index
print (vocab)
```

```python
X = []
for sentence in sentences:
    sentence = ViTokenizer.tokenize(sentence)
    words = sentence.split(" ")
    sentence_vec = np.zeros((100))
    for word in words:
        if word in vocab:
            sentence_vec+=w2v.key_to_index[word]
    X.append(sentence_vec)
```

```python
from sklearn.cluster import KMeans

n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters)
kmeans = kmeans.fit(X)
```

```python
from sklearn.metrics import pairwise_distances_argmin_min

avg = []
for j in range(n_clusters):
    idx = np.where(kmeans.labels_ == j)[0]
    avg.append(np.mean(idx))
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, X)
ordering = sorted(range(n_clusters), key=lambda k: avg[k])
summary = ' '.join([sentences[closest[idx]] for idx in ordering])
```
